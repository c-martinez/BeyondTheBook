```python
import pandas as pd
import pickle
```

```python
cKarina = pd.read_excel('../data/Book_selection.xlsx', 'CorpusOverview-Karina.csv')
cSanders = pd.read_excel('../data/Book_selection.xlsx', 'Corpus-Sanders')
corpora = { 'karina': cKarina, 
           'sanders': cSanders }
corporaDataDir = { 'karina': '../data/corpus.karina.nerINL.2012-10-26/', 
                  'sanders': '../data/corpus.sanders.nerINL.2012-10-24/' }
```

```python
for corpus in corpora.values():
    corpus['EntityCount'] = 0
    corpus['Mean_NL_Interest'] = 0
    
    # Named Entity Types:
    corpus['location'] = 0
    corpus['misc'] = 0
    corpus['organisation'] = 0
    corpus['person'] = 0
    corpus['Mean_NL_Interest_location'] = 0
    corpus['Mean_NL_Interest_misc'] = 0
    corpus['Mean_NL_Interest_organisation'] = 0
    corpus['Mean_NL_Interest_person'] = 0
```

```python
for corpusName, corpus in corpora.iteritems():
    for bookId in corpus['bookid']:
        data = pickle.load(open(corporaDataDir[corpusName] + bookId + '.pkl', 'r'))
        indexer = corpus['bookid']==bookId
        
        corpus.ix[indexer, 'EntityCount'] = data['Count'].sum()
        for wordType,typeData in data.groupby('Type'):
            corpus.ix[indexer, wordType] = typeData['Count'].sum() / data['Count'].sum()
            corpus.ix[indexer, 'Mean_NL_Interest_' + wordType] = (typeData['NLO'] * typeData['Count'] / typeData['Count'].sum()).sum()    
        
        corpus.ix[indexer, 'Mean_NL_Interest'] = (data['NLO'] * data['Count'] / data['Count'].sum()).sum()
```

```python
data = pd.concat(corpora.values())
# data[[ 'auteur', 'titel', 'location', 'misc', 'organisation', 'person', 'EntityCount', 'Mean_NL_Interest' ]]
data[[ 'auteur', 'titel', 'Mean_NL_Interest', 'Mean_NL_Interest_location', 'Mean_NL_Interest_misc', 'Mean_NL_Interest_organisation', 'Mean_NL_Interest_person' ]]
```

```python
columns=[ 'bookid', 'auteur', 'titel', 'EntityCount', 'location', 'misc', 'organisation', 'person', 
         'Mean_NL_Interest', 'Mean_NL_Interest_location', 'Mean_NL_Interest_misc', 
         'Mean_NL_Interest_organisation', 'Mean_NL_Interest_person'  ]

writer = pd.ExcelWriter('../data/TranslationsV3.xlsx')
data.to_excel(excel_writer=writer, sheet_name='Interest', columns=columns)
writer.save()
```
